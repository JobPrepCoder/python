# **int** (numeric)

we can use int data type to represent whole numbers(integral values)

note:--

in python 2 we have long data type for represeanting large number. but in python 3 we have only int.

```python
a = 90
type(a) #int
```

We can represeant int values in the following ways:---



1.   Decimal  ( 0---9)
2.   Octal (0----7)
3.   Binary (0--1)
4.   hexa decimal (0---15)



```python
# 1. Decimal base (10)

a=9
```

```python
#2. octal  base(8)
#0o 
#0O
```

```python
#decimal to octal

a = 89 # decimal
oct(a) #decimal to octal
```

```python
# octal to decimal(int)
v = 0o131 #octal
int(v) # octal to int
```

```python
# binary
#0b, 0B

a =9 # decimal
bin(a) #decimal to binary
```

```python
# binary to decimal
b = 0b1001 #binary
int(b) #binary to decimal
```

```python
c= 0o67 #octa
bin(c) #octa to binary
```

```python
k = 0b01000010 #binary
oct(k)# binary to oct
```

```python
#hexa

i =347 #decimal
hex(i) #decimal to hexa
```

```python
f = 8.99
type(f)
```

```python
g = 89E78
type(g)
```



1.   They can represnt values between the integers.

5.   floating point numbers are not convertable in binary,oct,hex like integers.




# **Complex Number**

A complex number is of the form 

a+bj

a--> Real part
bj--> imaginary part

a and b may be integer or float.



1.  in real part we can use int and other format like binary,oct,hexa.
2.   but in imaginary part we can use only float or int.



```python
k = 8 + 9j
type(k)
```

```python
l = 0x0010 + 8j
type(l)
```

```python
m = 7 + 0b110j
type(m)
```

Note:--

In some cases the exception overflow error is raised instead if the given number can not be represented through available
number of bytes.

# **string(str)**

It represents string data type.

It is a sequence of characters enclosed within single quotes or double quotes.

In python 3.x each charecter store in a string is a unicode character.

A string an hold any type of known charecters i.e. letters, numbers and special charcters.

```python
# string examples
'acaass'
"jjjjjjj"
"78899"
"@##$$%%^^^^&*+__))))?><"
```



1.   String in python are stored by storing each charcter separately in  contiguous locations.
2.   The charcters of the string are given in two way indexes.


*   in forward direction(0,1,2----------)
*   in backword directon(-1,-2,-3,-4-------)

3. First charcter of the string is at index 0 or -length.
4. Second charcter of the string is at index 1 or -(length -1)
5. Second last charcter of the string is at index (length-2) or -2.
6. Last charcter of the string is at index (length -1) or -1
